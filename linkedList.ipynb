```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedList:
    
    def __init__(self):
        self.head = None
    
    def printList(self):
        temp = self.head
        print('HEAD', end=" -> ")
        while(temp):
            print(temp.data, end=" -> ")
            temp = temp.next
        print("NULL")

    def insertAtStart(self,node):
        
        if( not self.head):
            # self head is empty
            self.head = node
            return 
        
        temp = self.head
        self.head = node
        node.next = temp

    
#created a new and empty linked list
llist = LinkedList() 

# creating new nodes
first = Node(1) 
second = Node(2)
third = Node(3)

# connecting nodes

llist.head = first
llist.head.next = second
second.next = third

# printing the Linked list

llist.printList()
```
